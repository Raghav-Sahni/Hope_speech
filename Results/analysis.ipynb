```python
import pandas as pd
import os
```

```python
parent_dir = os.path.join(os.path.abspath(os.path.join(os.getcwd(), os.pardir)))
```

```python
from operator import index


df = pd.read_csv(os.path.join(parent_dir, 'Results\MidEval_Reported_Scores.csv'), index_col=0)
```

```python
df['Algorithm'] = df['Classifier'].str.split('_').str[:-1].str.join('_')
df['Embedding'] = df['Classifier'].str.split('_').str[-1]
```

```python
df[df['Algorithm'] == 'Linear_Discriminant_Analysis']
```

```python
# For Every Unique Algorithm keep row with max Weighted F1 Test
best_results = df.sort_values('Weighted F1 Test', ascending=False).drop_duplicates(subset=['Algorithm'], keep='first')
```

```python
best_results
```

```python
best_results.to_csv(os.path.join(parent_dir, 'Results\Best_Results_Pre_MidEval.csv'))
```

```python
df_showcase = best_results[['Algorithm', 'Embedding', 'Weighted F1 Test']]
```

```python
df_showcase.to_csv(os.path.join(parent_dir, 'Results\Best_Results_Pre_MidEval_Showcase.csv'))
```

```python
# Save to an image png
df_showcase.to_html(os.path.join(parent_dir, 'Results\Best_Results_Pre_MidEval.html'))

# Save to an image png
```

```python
df_lda = df[df['Algorithm'] == 'Linear_Discriminant_Analysis']
df_lda = df_lda.sort_values('Weighted F1 Test', ascending=False)
df_lda[['Algorithm', 'Embedding', 'Weighted F1 Test']].to_csv(os.path.join(parent_dir, 'Results\Best_Results_Pre_MidEval_LDA.csv'))
```
